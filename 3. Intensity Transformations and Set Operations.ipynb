# Intensity Transformations and Set Operations


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

### Negative Transformation



```python
image = cv.imread('image3.jpeg', cv.IMREAD_GRAYSCALE)
```

```python
image.shape
```

```python
negative_image = 255 - image

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(negative_image, cmap='gray')
plt.title("Negative Image")
plt.axis('off')

plt.show()
```

### Log Transformation

```python
c = 255 / np.log(1 + np.max(image))  # np.log(1 + max(image)) computes the maximum possible log-transformed value (ensure betwwen 0-255)
log_image = c * np.log(1 + image)

log_image = np.array(log_image, dtype=np.uint8) # uint8 (Unsigned 8-bit integer) is the standard datatype for grayscale and color images

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(log_image, cmap='gray')
plt.title("Log Transformed Image")
plt.axis('off')

plt.show()
```

### Power-Law Transform

```python
image_norm = image / 255.0 # Normalize betwwen 0 and 1

gamma = 2.2  # if gamma>1, darker and if gamma<1, brighter)
gamma_image = np.power(image_norm, gamma)
gamma_image = np.uint8(gamma_image * 255) # Rescale to 8 bits

gamma2 = 0.5
gamma_image2 = np.power(image_norm, gamma)
gamma_image2 = np.uint8(gamma_image * 255)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gamma_image, cmap='gray')
plt.title(f"Gamma Transformed Image (γ={gamma})")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(gamma_image2, cmap='gray')
plt.title(f"Gamma Transformed Image (γ={gamma2})")
plt.axis('off')

plt.show()
```

## Set Operations

```python
image1 = cv.imread('image5.jpg', cv.IMREAD_GRAYSCALE)
image2 = cv.imread('image6.jpg', cv.IMREAD_GRAYSCALE)
```

### Union

```python
union_image = cv.bitwise_or(image1, image2)

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(image1, cmap='gray')
plt.title("Image 1")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(image2, cmap='gray')
plt.title("Image 2")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(union_image, cmap='gray')
plt.title("Union (A ∪ B)")
plt.axis('off')

plt.show()
```

### Intersection


```python
intersection_image = cv.bitwise_and(image1, image2)

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(image1, cmap='gray')
plt.title("Image 1")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(image2, cmap='gray')
plt.title("Image 2")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(intersection_image, cmap='gray')
plt.title("Intersection (A ∩ B)")
plt.axis('off')
plt.show()
```

### Difference

```python
difference_image = cv.bitwise_and(image1, cv.bitwise_not(image2))

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(image1, cmap='gray')
plt.title("Image 1")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(image2, cmap='gray')
plt.title("Image 2")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(difference_image, cmap='gray')
plt.title("Set Difference (A - B)")
plt.axis('off')
plt.show()
```

### XOR

```python
xor_image = cv.bitwise_xor(image1, image2)

plt.figure(figsize=(10,5))

plt.subplot(1,3,1)
plt.imshow(image1, cmap='gray')
plt.title("Image 1")
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(image2, cmap='gray')
plt.title("Image 2")
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(xor_image, cmap='gray')
plt.title("XOR (A, B)")
plt.axis('off')

plt.show()
```

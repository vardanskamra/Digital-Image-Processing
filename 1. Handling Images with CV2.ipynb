# Basic Image Handling and Processing

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import sys
```

### Reading an Image

```python
img = cv.imread("image2.jpg")
if img is None:
    sys.exit("Could not read the image.")

cv.imshow("Display window", img)
k = cv.waitKey(0) # Wait indefinitely for a key press (takes time in ms)

if k == ord("s"): # To handle specific events
    cv.imwrite("hello.png", img)

cv.destroyAllWindows()
```

```python
img.shape
```

### Using Matplotlib to Show an Image

```python
plt.imshow(img) #cv2 uses BGR color format and matplotlib uses RGB color format
```

### Converting BGR color to RGB color format

```python
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

### Converting an Image to Greyscale

```python
img = cv.imread("image2.jpg")
grey_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

"""
cv.imshow('grey image', grey_img)
cv.waitKey(0)
cv.destroyAllWindows()
"""

plt.imshow(grey_img, cmap="gray")
img.shape
```

### Resizing an Image

```python
img = cv.imread("image2.jpg") #(800 X 800 Image)
new_width, new_height = 200, 400  # New dimensions
resized_image = cv.resize(img, (new_width, new_height))
print(resized_image.shape)

plt.imshow(cv.cvtColor(resized_image, cv.COLOR_BGR2RGB))
```

### Save Image in Another Format

```python
img = cv.imread('image2.jpg')
cv.imwrite('image2.png', img)
```

### Save with Different Quality


```python
image = cv.imread('image2.png')
# Save with quality parameter (0-100)
cv.imwrite('new_image2.jpg', image, [cv.IMWRITE_JPEG_QUALITY, 50])
```

```python
cap = cv.VideoCapture('video1.mp4')

if (cap.isOpened()== False):
    print("Error opening video file")

# Read until video is completed
while(cap.isOpened()):

# Capture frame-by-frame
    ret, frame = cap.read()
    if ret == True:
    # Display the resulting frame
        cv.imshow('Frame', frame)

    # Press Q on keyboard to exit
        if cv.waitKey(25) & 0xFF == ord('q'):
            break

# Break the loop
    else:
        break

# When everything done, release
# the video capture object
cap.release()
```
